```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
costumer_data = pd.read_csv('hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv')
```

```python
costumer_data.head()
```

```python
costumer_data.shape
```

```python
data = costumer_data.iloc[:,3:5].values
```

```python
data
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10,7))
plt.title('Costumer Dendograms')
dend = shc.dendrogram(shc.linkage(data , method='ward'))
```

```python
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster = AgglomerativeClustering(n_clusters=5, linkage='average')
labels_ = cluster.fit_predict(data)
```

```python
labels_
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data[:,0] , data[:,1],c = cluster.labels_ , cmap='rainbow')
```
