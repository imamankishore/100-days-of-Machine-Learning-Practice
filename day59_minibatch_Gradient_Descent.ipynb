```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
X_train , X_test , y_train ,y_test = train_test_split(X,y,
                                                     test_size=0.2,
                                                     random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr = lr.fit(X_train , y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
import random
```

```python
class MBGDRegressor:
    def __init__(self, batch_size, learning_rate, epochs):  # Fixed: Changed from 'self' to '__init__'
        self.coef_ = None
        self.intercept_ = None
        self.batch_size = batch_size
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self, X_train, y_train):
        # init the coef and intercept
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])

        for i in range(self.epochs):
            for j in range(int(X_train.shape[0]/self.batch_size)):
                idx = random.sample(range(X_train.shape[0]), self.batch_size)  # Fixed: random.sample needs a sequence

                y_hat = np.dot(X_train[idx], self.coef_) + self.intercept_
                intercept_der = -2 * np.mean(y_train[idx] - y_hat)  # Fixed: y-train to y_train
                self.intercept_ = self.intercept_ - (self.lr * intercept_der)

                coef_der = -2 * np.dot((y_train[idx] - y_hat), X_train[idx])
                self.coef_ = self.coef_ - (self.lr * coef_der)

            print(self.intercept_, self.coef_) 
            
    def predict(self, X_test):  
        return np.dot(X_test, self.coef_) + self.intercept_
```

```python
reg =MBGDRegressor(batch_size = 10 , learning_rate=0.1, epochs=50)
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```
