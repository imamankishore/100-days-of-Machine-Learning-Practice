```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
X , y = make_regression(n_samples=100,n_features= 1 , n_informative=1 , n_targets=1, noise=20 , random_state=13)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
from sklearn.linear_model import Ridge
```

```python
r = Ridge(alpha=10)
```

```python
r = r.fit(X,y)
```

```python
r.coef_
```

```python
r.intercept_
```

```python
r1 = Ridge(alpha=100)
```

```python
r1.fit(X,y)
```

```python
r1.coef_
```

```python
r1.intercept_
```

```python
plt.plot(X,y , 'b.')
plt.plot(X, lr.predict(X) , color='red' , label = 'alpha0')
plt.plot(X , r.predict(X) , color='green' , label = 'alpha10')
plt.plot(X , r1.predict(X) , color='yellow' , label= 'alpha100')
plt.legend()
```
