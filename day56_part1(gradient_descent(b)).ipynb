```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X , y = make_regression(n_samples = 4 , n_features=1 , n_informative=1 , n_targets=1 , noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
```

```python
#apllying OLS method(Ordinary Least Square)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X) , color='red')
```

```python
#lets aplly Gradient Descent assuming slope is constant(m) = 78.35
#and let's assume starting value for intercept b = 0
```

```python
y_pred = ((78.35 * X) + 0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X) , color='red', label = 'OLS')
plt.plot(X, y_pred , color='blue' , label = 'b=0')
plt.legend()
```

```python
m = 78.35
b=0

loss_slope = -2* np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
#lets take learning rate = 0.1
lr = 0.1

step_size = loss_slope*lr
step_size
```

```python
#calculating the new intercept
b = b -step_size
b
```

```python
y_pred1 = ((78.35*X) + b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X) , color='red', label = 'OLS')
plt.plot(X, y_pred , color='blue' , label = 'b=0')
plt.plot(X,y_pred1 , color = 'skyblue' , label=f'b={b.round(2)}')
plt.legend()
```

```python
#iteration 2
```

```python
loss_slope = -2* np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b -step_size
b
```

```python
y_pred2 = ((78.35*X) + b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X) , color='red', label = 'OLS')
plt.plot(X, y_pred , color='blue' , label = 'b=0')
plt.plot(X,y_pred1 , color = 'skyblue' , label=f'b={b.round(2)}')
plt.plot(X,y_pred2 , color='yellow' , label=f'b={b.round(2)}')
plt.legend()
```

```python
#Iteration2
```

```python
loss_slope = -2* np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b -step_size
b
```

```python
y_pred3 = ((78.35*X) + b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X) , color='red', label = 'OLS')
plt.plot(X, y_pred , color='blue' , label = 'b=0')
plt.plot(X,y_pred1 , color = 'skyblue' , label=f'b={b.round(2)}')
plt.plot(X,y_pred2 , color='yellow' , label=f'b={b.round(2)}')
plt.plot(X,y_pred3 , color='pink' , label = f'b={b.round(2)}')
plt.legend()
```

```python
b =  -100
m= 78.35
lr=0.1

epochs = 10

for i in range(epochs):
    loss_slope = -2 *np.sum(y- m*X.ravel() -b)
    b = b-(lr * loss_slope)

    y_pred = m*X + b

    plt.plot(X,y_pred)


plt.scatter(X,y)
```
